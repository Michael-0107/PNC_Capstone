```python
import re
```

```python
def process_line(line):
    return ' '.join(line.split()[2:-4])

def process_file(input_file, output_file):
    with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
        for line in infile:
            processed_line = process_line(line.strip())
            outfile.write(processed_line + '\n')


input_file = 'name/ConsumerStaples.txt'
output_file = 'processed_data_1.txt'


process_file(input_file, output_file)

print(f'Finished {output_file}')
```

```python
def process_company_name(name):
    # 将公司名称转换为小写
    name = name.lower()
    # 将符号替换为空格
    name = re.sub(r'[^\w\s]', ' ', name)
    # 将多个连续的空格替换为单个空格
    name = re.sub(r'\s+', ' ', name).strip()
    return name
```

```python
def process_file(input_file, output_file):
    with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
        for line in infile:
            # 处理每一行
            processed_name = process_company_name(line.strip())
            # 将处理后的公司名称写入输出文件
            outfile.write(processed_name + '\n')
```

```python
input_file = 'data/companies.txt'
output_file = 'data/processed_companies.txt'

# 处理文件
process_file(input_file, output_file)

print(f'处理完成，结果已保存到 {output_file}')
```

```python
def generate_sql_query(companies):
    base_query = """
SELECT
  DATE,
  Organizations,
  DocumentIdentifier
FROM 
  `gdelt-bq.gdeltv2.gkg_partitioned`
WHERE
  TIMESTAMP_TRUNC(_PARTITIONTIME, DAY) BETWEEN TIMESTAMP("2009-09-01") AND TIMESTAMP("2024-04-01")
  AND (
    {}
  )
"""

    conditions = []
    for company in companies:
        company = company.strip().lower()
        condition = f"REGEXP_CONTAINS(Organizations, r'\\b{company}\\b')"
        conditions.append(condition)
    
    conditions_str = " OR \n    ".join(conditions)
    query = base_query.format(conditions_str)
    return query

def read_companies(file_path):
    with open(file_path, 'r') as file:
        companies = file.readlines()
    return companies

# 指定公司名称文件
input_file = 'data/companies.txt'

# 读取公司名称
companies = read_companies(input_file)

# 生成 SQL 查询
sql_query = generate_sql_query(companies)

# 打印生成的 SQL 查询
print(sql_query)
```
